```python
from pycaret.regression import *

s = setup(data = df_clean, 
          categorical_features=['type', 'technology', 'ftth_takeover'],
          target = 'total_cost', 
          session_id=123,
          ignore_features='mdu_name',
          remove_outliers=True,
          #outliers_threshold=0.95,
          normalize=True,
          normalize_method='robust',
          )
```
